```python
lst = [[3,2,1,8],[50,3,7,9],[2,3,7,8]]
ans=[]
for i in range(len(lst)):
    for j in range(len(lst[i])):
        
        if lst[i][j] % 2 == 0:   # even
            lst[i][j] -= 1
        else:                    # odd
            lst[i][j] += 1

print(lst)
```
