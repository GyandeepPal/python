```python
color1 = ["red","orange","blue","white"]
color2 = ["black","yellow","green","blue"]

lst1 = []
lst2 = []

# elements in color1 but not in color2
for i in color1:
    if i not in color2:
        lst1.append(i)

# elements in color2 but not in color1
for i in color2:
    if i not in color1:
        lst2.append(i)

print(lst1)
print(lst2)
```

DuplicateList

```python
lst = [0,0,0,4,0,3,0,9,5,3,8,2,8,4,8,8,5,8,3,9,2,9,4]

seen = set()
duplicates = set()

for i in lst:
    if i in seen:
        duplicates.add(i)
    else:
        seen.add(i)

print(list(duplicates))
```
